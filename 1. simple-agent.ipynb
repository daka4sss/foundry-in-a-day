### ライブラリ読込み & 設定 & クライアント準備

```python
import os
from azure.identity import DefaultAzureCredential
from azure.ai.projects import AIProjectClient
from azure.ai.projects.models import PromptAgentDefinition
from dotenv import load_dotenv

# .env 読み込み
load_dotenv()

# Azure AI Projects client
project_client = AIProjectClient(
    endpoint=os.environ["project_endpoint"], # Foundryプロジェクトのエンドポイント
    credential=DefaultAzureCredential() # Azure認証（推奨）
)
```

### エージェントを作成

```python
agent_name = "simple-agent-vs"
model_deployment_name = "gpt-4.1"

agent = project_client.agents.create_version(
    agent_name=agent_name,
    definition=PromptAgentDefinition(
        model=model_deployment_name,
        instructions="情報を見つけるのに役立つAIアシスタントです。関西弁で回答してください。",
    ),
)

print(f"✅ エージェント作成完了 (名前： {agent.name}, バージョン: {agent.version})")
```

### プロンプト入力＆実行

```python
# OpenAIクライアントの取得
openai_client = project_client.get_openai_client()

# ここを毎回自由に書き換えて試す！
user_prompt = "京都のおすすめ観光スポットを100文前後字教えて！"

response = openai_client.responses.create(
    input=[{"role": "user", "content": user_prompt}],
    extra_body={"agent": {"name": agent.name, "type": "agent_reference"}},
)

print("=== レスポンス ===")
print(response.output_text)
```
